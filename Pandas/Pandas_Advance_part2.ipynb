# to drop column or rows from data frame

 #### you have to give axis no. the,  column/ row not drop permananatly ... it drop for exicution time only
 
 #### for permanat delet use, --   __inplace=True__

```python
df.drop('PassengerId')
```

```python
df.drop('PassengerId', axis=1) # this will not drop permanatally
```

```python
df.drop('PassengerId', axis=1, inplace=True)  # this will drop permanatally
```
