# marge and join

```python
data1=pd.DataFrame({"key1":[2,3,4,5,8],"key2":[5,8,9,6,10],"key3":[9,6,3,7,9]})
```

```python
data2=pd.DataFrame({"key1":[2,33,4,55,8],"key4":[35,78,29,76,90],"key5":[19,26,43,57,69]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2, how='left')
```

```python
pd.merge(data1,data2, how='right') # all the combination
```

```python
pd.merge(data1,data2, how='outer') 
```

```python
pd.merge(data1,data2, how='left', on='key1') 
```

```python
pd.merge(data1,data2, how='left', left_on='key2' , right_on='key4') 
```
