## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

## 1. Linear Interpolation

```python
# create two array
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
# plote those data points
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
## interpolate the data using linear interpolation
x_new=np.linspace(1,5,10) ## create new x values array , x range is x=np.array([1,2,3,4,5]) --> 1-5 within and 10 point create
y_interp=np.interp(x_new,x,y) # One-dimensional linear interpolation for monotonically increasing sample points.
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp) # within 1-5 range more data set is created
```

## 2. Cubic Interpolation With Scipy

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
##create a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
# interpolate the data
x_new = np.linspace(1, 5, 10)
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

## 3. Polynomial Interpolation

```python
import numpy as np

# create some sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
```

```python
# interpolate the data using polynomial interpolation
p=np.polyfit(x,y,2) #degree of polynomial =2

# if degree of polynomial=1 ==> leaniar
# if degree of polynomial = 2 ==> exponential 
# if degree of polynomial = 3 ==> more exponential
```

```python
x_new = np.linspace(1, 5, 10) # create new x values w.r.t 10  data points added within the same privious range 
y_interp = np.polyval(p, x_new) # interpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
