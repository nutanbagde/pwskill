# 1. scatter Plot

```python
plt.scatter(x,y)   # graph type 
plt.xlabel('x_axis')  # Set the label for the x-axis.
plt.ylabel('y_axis')  #Set the label for the x-axis.
plt.title("This is my x vs y")  # Set a title for the Axes.
```

### 1.1 c= HTML color code google it and select the name of color

```python
plt.scatter(x,y, c= '#0CCB4C')   # c= HTML color code google it and select the name of color
plt.xlabel('x_axis')  
plt.ylabel('y_axis')  
plt.title("This is my x vs y") 
```

### 1.2. alpha=(ranges 0 to 1) it controll the intencity of color

```python
plt.scatter(x,y, c= '#0CCB4C', alpha=0.5)   #  alpha=0.5  The alpha blending value, between 0 (transparent) and 1 (opaque).
plt.xlabel('x_axis')  
plt.ylabel('y_axis')  
plt.title("This is my x vs y") 
```

### 1.3. marker= how to represent the point where circular or x

```python
plt.scatter(x,y, c= '#0CCB4C', alpha=0.5, marker='x')   
plt.xlabel('x_axis')  
plt.ylabel('y_axis')  
plt.title("This is my x vs y") 
```

### 1.4. plt.grid() : Configure the grid lines.

```python
plt.scatter(x,y, c='#0CCB4C', alpha=0.9, marker='x')
plt.xlabel('x_axis')  
plt.ylabel('y_axis')  
plt.title("This is my x vs y") 
plt.grid()
```

### 1.5. change the figure size: 
* plt.figure(figsize=(length,hight))  

```python
plt.figure(figsize=(6,2))

plt.scatter(x,y, c= '#0CCB4C' , alpha= 0.9)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title('"This is my x vs y"')
```

# 3. Bar plot.

### 3.1. FigureSize, Xlabel, ylabel, title and grid

# 6. 3-D Plot

```python
# create data points
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)

# create a figure and save it in fig variable
fig=plt.figure()

# create an axis of 3D plot using fig and save in ax variable
ax=fig.add_subplot(projection='3d')

# project your points on that axis
ax.scatter(x,y,z, c='#700EBE')

# label the axis
plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.show()
```
